```python
def load(file_name) -> list:
    output=[]
    with open(file_name) as f:
        data = f.readlines()
    for line in data:
        line = line.replace('\n', '').split(' -> ')
        line = [line[0].split(','), line[1].split(',')]
        line = [tuple([int(i) for i in l]) for l in line]
        output.append(line)
    return output

if __name__ == '__main__':
    starts_stops = load(r'data\day_5.txt')
```

```python
def give_path_for_all(starts_stops) -> list:
    paths = []
    for start_stop in starts_stops:
        paths.append(give_path(start_stop))
    return paths

def give_path(start_stop) -> list:
    start = start_stop[0]
    stop = start_stop[1]
    if len(start_stop)>2: raise ValueError
    if start[0] == stop[0] or start[1] == stop[1]:
        path = 'yep'
        if start[0] == stop[0]:
            pos_min = min(start[1], stop[1])
            pos_max = max(start[1], stop[1])
            path = [[start[0], y] for y in range(pos_min, pos_max+1)]
            
        if start[1] == stop[1]:
            pos_min = min(start[0], stop[0])
            pos_max = max(start[0], stop[0])
            path = [[x, start[1]] for x in range(pos_min, pos_max+1)]
            
    else: 
        path = []
        x_increment = -int(abs(start[0]-stop[0])/(start[0]-stop[0]))
        y_increment = -int(abs(start[1]-stop[1])/(start[1]-stop[1]))
        x_ = list(range(start[0], stop[0]+x_increment, x_increment))
        y_ = list(range(start[1], stop[1]+y_increment, y_increment))
        for x, y in zip(x_,y_):
            path.append([x, y])
    return path
```

```python
grid = [[0 for j in range(0, 1000)] for i in range(0, 1000)]

paths = give_path_for_all(load(r'data\day_5.txt'))
for i, path in enumerate(paths):
    if path != [[], []]:
        for pos in path:
            grid[pos[0]][pos[1]]+=1
```
